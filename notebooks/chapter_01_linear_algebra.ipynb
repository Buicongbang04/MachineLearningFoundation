## 1. Setup

```python
import numpy as np

SEED = 42
rng = np.random.default_rng(SEED)
np.set_printoptions(precision=4, suppress=True)
```

## 2. Vectors and matrices

A vector is a 1-D array; a matrix is a 2-D array. `.shape` tells you which is which.

```python
x = np.array([1.0, 2.0, 3.0])
X = np.array([[1, 2, 3],
              [4, 5, 6]])
print('x shape =', x.shape)
print('X shape =', X.shape)
print('X[0, :] =', X[0, :])     # first row
print('X[:, 1] =', X[:, 1])     # second column
```

## 3. Transpose

`X.T` swaps rows and columns. A column vector `(n,1)` transposes to a row vector `(1,n)`.

```python
print('X      shape =', X.shape)
print('X.T    shape =', X.T.shape)
print('X.T:')
print(X.T)
```

## 4. Inner product (dot product)

For $\mathbf{x}, \mathbf{y} \in \mathbb{R}^n$: $\mathbf{x}^\top \mathbf{y} = \sum_i x_i y_i$. Geometrically this is $\|\mathbf{x}\|\|\mathbf{y}\| \cos\theta$.

```python
x = np.array([1.0, 2.0, 3.0])
y = np.array([4.0, 5.0, 6.0])
print('np.dot(x, y) =', np.dot(x, y))
print('x @ y        =', x @ y)
print('manual sum   =', sum(xi * yi for xi, yi in zip(x, y)))
```

## 5. Matrix multiplication

$\mathbf{C} = \mathbf{A}\mathbf{B}$ where $\mathbf{A}$ is $m \times n$, $\mathbf{B}$ is $n \times p$, $\mathbf{C}$ is $m \times p$. The $(i, j)$ entry of $\mathbf{C}$ is the dot product of row $i$ of $\mathbf{A}$ with column $j$ of $\mathbf{B}$.

```python
A = np.array([[1, 2],
              [3, 4],
              [5, 6]])      # 3 x 2
B = np.array([[7,  8,  9],
              [10, 11, 12]])# 2 x 3
C = A @ B                   # 3 x 3
print('A shape =', A.shape, '  B shape =', B.shape, '  C shape =', C.shape)
print('C =')
print(C)
```

## 6. Linear model: matrix-vector product

A dataset $\mathbf{X} \in \mathbb{R}^{N \times d}$ has one sample per row. The linear-model prediction across the whole dataset is
$$\hat{\mathbf{y}} = \mathbf{X}\mathbf{w} + b\mathbf{1}.$$

```python
N, d = 5, 3
X = rng.standard_normal(size=(N, d))
w = np.array([0.5, -0.2, 1.0])
b = 0.1
y_hat = X @ w + b
print('X shape =', X.shape, '  w shape =', w.shape, '  y_hat shape =', y_hat.shape)
print('y_hat =', y_hat)
```

## 7. Norms

- $\ell_1$ — sum of absolute values.
- $\ell_2$ — square-root of sum of squares (Euclidean length).
- $\ell_\infty$ — largest absolute value.

```python
x = np.array([3.0, -4.0, 12.0])
print('L1 (manhattan)   :', np.linalg.norm(x, ord=1))
print('L2 (euclidean)   :', np.linalg.norm(x, ord=2))
print('Linf (max-abs)   :', np.linalg.norm(x, ord=np.inf))
```

## 8. Special matrices: identity and inverse

```python
I = np.eye(3)
A = np.array([[4.0, 2.0],
              [1.0, 3.0]])
A_inv = np.linalg.inv(A)
print('I:\n', I)
print('A @ A_inv:\n', A @ A_inv)   # should be I
```

## 9. Eigenvalues and eigenvectors (intuition)

$\mathbf{A}\mathbf{v} = \lambda\mathbf{v}$: the matrix only scales $\mathbf{v}$ by the scalar $\lambda$, no rotation. These show up again in PCA and SVD (later chapters).

```python
A = np.array([[2.0, 0.0],
              [0.0, 3.0]])
eigvals, eigvecs = np.linalg.eig(A)
print('eigvals =', eigvals)
print('eigvecs (columns):\n', eigvecs)
```

## 10. CHECKPOINT — implement the four building blocks

Most of this course's algorithms boil down to four primitives. Implement each twice (NumPy + by hand) and compare.

### 10.1 Euclidean distance

```python
def euclidean_distance(a: np.ndarray, b: np.ndarray) -> float:
    return float(np.sqrt(np.sum((a - b) ** 2)))

a = np.array([1.0, 2.0, 3.0])
b = np.array([4.0, 6.0, 3.0])
print('manual      :', euclidean_distance(a, b))
print('np.linalg   :', np.linalg.norm(a - b))
```

### 10.2 Cosine similarity

```python
def cosine_similarity(a: np.ndarray, b: np.ndarray) -> float:
    return float(a @ b / (np.linalg.norm(a) * np.linalg.norm(b)))

u = np.array([1.0, 0.0, 0.0])
v = np.array([1.0, 1.0, 0.0])
w = np.array([-1.0, 0.0, 0.0])
print('cos(u, v)   =', cosine_similarity(u, v))   # 0.7071
print('cos(u, u)   =', cosine_similarity(u, u))   # 1.0
print('cos(u, w)   =', cosine_similarity(u, w))   # -1.0
```

### 10.3 Matrix-vector multiplication, by hand

```python
def matvec(A: np.ndarray, x: np.ndarray) -> np.ndarray:
    m, n = A.shape
    assert x.shape == (n,), 'shape mismatch'
    out = np.zeros(m)
    for i in range(m):
        for j in range(n):
            out[i] += A[i, j] * x[j]
    return out

A = rng.standard_normal(size=(4, 3))
x = rng.standard_normal(size=(3,))
print('manual:', matvec(A, x))
print('A @ x :', A @ x)
assert np.allclose(matvec(A, x), A @ x)
```

### 10.4 L2 normalization

Mapping a vector to its unit-length version $\hat{\mathbf{x}} = \mathbf{x} / \|\mathbf{x}\|_2$ is a standard preprocessing step before computing cosine similarity in batch.

```python
def l2_normalize(X: np.ndarray, axis: int = -1, eps: float = 1e-12) -> np.ndarray:
    norms = np.linalg.norm(X, axis=axis, keepdims=True)
    return X / np.maximum(norms, eps)

X = rng.standard_normal(size=(5, 3))
Xn = l2_normalize(X, axis=1)
print('row norms after normalization:', np.linalg.norm(Xn, axis=1))
```

## 11. Summary

- A dataset is a matrix `X` of shape `(N, d)`. A linear model is `X @ w + b`.
- Dot products show up everywhere — predictions, similarities, gradient updates.
- L2 distance measures magnitude differences; cosine similarity measures angle only.
- L2 normalization makes rows comparable by angle.
